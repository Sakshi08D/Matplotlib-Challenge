```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Read data files
mouse_metadata = pd.read_csv('data/Mouse_metadata.csv')
study_results = pd.read_csv('data/Study_results.csv')

# Merge DataFrames
merged_df = pd.merge(mouse_metadata, study_results, on='Mouse ID')


# Display the data table for preview
print(merged_df.head(10)) 
```

```python
# Display number of unique mice IDs
unique_mice = len(merged_df['Mouse ID'].unique())
print("Number of unique mice:", unique_mice)
```

```python
# Check for duplicate mouse IDs with the same time points
duplicates = merged_df[merged_df.duplicated(subset=['Mouse ID', 'Timepoint'])]
```

```python
# Display the data associated with duplicate mouse IDs
print("Duplicate mouse IDs with same time points:")
print(duplicates)
```

```python
# Create a new DataFrame with duplicates removed
cleaned_df = merged_df.drop_duplicates(subset=['Mouse ID', 'Timepoint'])
```

```python
# Display updated number of unique mice IDs
print("Updated number of unique mice:", cleaned_df['Mouse ID'].nunique())
```

## Summary Statistics

```python
import numpy as np

# Generate Summary Statistics
summary_stats = cleaned_df.groupby('Drug Regimen').agg(
    Mean=('Tumor Volume (mm3)', np.mean),
    Median=('Tumor Volume (mm3)', np.median),
    Variance=('Tumor Volume (mm3)', np.var),
    Standard_Deviation=('Tumor Volume (mm3)', np.std),
    SEM=('Tumor Volume (mm3)', st.sem)
)
print(summary_stats)
```

```python
# Generate Summary Statistics using the aggregation method
summary_stats = cleaned_df.groupby('Drug Regimen').agg(
    Mean=('Tumor Volume (mm3)', np.mean),
    Median=('Tumor Volume (mm3)', np.median),
    Variance=('Tumor Volume (mm3)', np.var),
    Standard_Deviation=('Tumor Volume (mm3)', np.std),
    SEM=('Tumor Volume (mm3)', st.sem)
)

summary_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate Quartiles, Find Outliers, and Create a Box Plot
treatment_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
tumor_vol_data = []
```

```python
# Create grouped DataFrame
last_timepoint = cleaned_df.groupby('Mouse ID')['Timepoint'].max()
last_timepoint_df = pd.DataFrame(last_timepoint)
merged_last_timepoint = pd.merge(last_timepoint_df, cleaned_df, on=['Mouse ID', 'Timepoint'], how='left')
```

```python
# Loop through treatments
for treatment in treatment_list:
    final_tumor_vol = merged_last_timepoint.loc[merged_last_timepoint['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']
    tumor_vol_data.append(final_tumor_vol)
```

```python
# Calculate quartiles and IQR
quartiles = final_tumor_vol.quantile([0.25, 0.5, 0.75])
lower_q = quartiles[0.25]
upper_q = quartiles[0.75]
iqr = upper_q - lower_q
```

```python
# Find outliers
lower_bound = lower_q - (1.5 * iqr)
upper_bound = upper_q + (1.5 * iqr)
outliers = final_tumor_vol[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    
print(f"{treatment} outliers: {outliers}")
```

```python
# Create a box plot
fig, ax = plt.subplots()
ax.boxplot(tumor_vol_data, sym='r.')
ax.set_title('Final Tumor Volume Across Treatment Regimens')
ax.set_xlabel('Drug Regimen')
ax.set_ylabel('Tumor Volume (mm3)')
ax.set_xticklabels(treatment_list)
plt.show()
```

## Line and Scatter Plots

```python
# Create a Line Plot
capomulin_df = cleaned_df.loc[cleaned_df['Drug Regimen'] == 'Capomulin']
mouse_id = capomulin_df['Mouse ID'].iloc[0]
mouse_data = capomulin_df.loc[capomulin_df['Mouse ID'] == mouse_id]
plt.plot(mouse_data['Timepoint'], mouse_data['Tumor Volume (mm3)'])
plt.title(f'Capomulin Treatment for Mouse {mouse_id}')
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Create a Scatter Plot
capomulin_avg = capomulin_df.groupby('Mouse ID').mean()
plt.scatter(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])
plt.title('Tumor Volume vs Mouse Weight for Capomulin Treatment')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate Correlation and Regression
correlation = st.pearsonr(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])[0]
print(f"Correlation between mouse weight and average tumor volume: {correlation}")

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])
regress_values = capomulin_avg['Weight (g)'] * slope + intercept

plt.scatter(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])
plt.plot(capomulin_avg['Weight (g)'], regress_values, color='red')
plt.title('Tumor Volume vs Mouse Weight for Capomulin Treatment')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```
